```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/mpg.csv")
```

```python
df.head()
```

```python
df = df.drop("name", axis=1
             )
```

```python
df.head()
```

```python
df.origin.value_counts()
```

```python
#transofrmamos en numeros
```

```python
def origin(x):
    if x== "usa": 
        return 1
    else: 
        return 0
```

```python
df["origin"] = df.origin.apply(origin)
```

```python
df.head()
```

```python
X = df.drop("mpg", axis=1)
y = df.mpg
```

```python
pip install scikit-learn
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV
from sklearn.pipeline import Pipeline
```

```python
norm = MinMaxScaler()
norm.fit(X)
X_sc = norm.transform(X)
```

```python
params = {"max_depth": [3,5,7,9], "n_estimators":[100,200,300]}
rf = RandomForestRegressor()

grid = RandomizedSearchCV(
    estimator=rf,
    param_distributions=params,
    n_iter= 200,
    cv=3,
    n_jobs=-1
)

grid.fit(X_sc,y)
```

```python
grid.best_score_
```

```python
grid.best_estimator_.score(X_sc,y)
```

```python
#ya tenemos el modelo, lo guardamos
```

```python
import pickle
with open("model.pkl", "wb") as f:
    pickle.dump(grid.best_estimator_, f)
```

```python
with open("model.pkl", "rb") as f:
    saved_model = pickle.load(f)
```

```python
pipe = Pipeline([("norm", MinMaxScaler()),( "rf", RandomForestRegressor(n_jobs=-1))])

params = {"max_depth": [3,5,7,9], "n_estimators":[100,200,300]}

grid = RandomizedSearchCV(
    estimator=pipe,
    param_distributions=params,
    n_iter= 200,
    cv=3,
    n_jobs=-1
)

grid.fit(X,y)
```
